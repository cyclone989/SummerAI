```python
ram_price = pd.read_csv('datasets/ram_price.csv')
ram_price
```

```python
plt.yticks(fontname='Arial')
plt.semilogy(ram_price.date, ram_price.price)
plt.xlabel("년")
plt.ylabel("가격 ($/Mbyte)")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
data_train = ram_price[ram_price.date < 2000]
data_test = ram_price[ram_price.date >= 2000]
X_train = data_train.date[:, np.newaxis] #2차원 데이터로 변경
y_train = np.log(data_train.price)

tree = DecisionTreeRegressor().fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)

X_all = ram_price.date[:, np.newaxis]
pred_tree = tree.predict(X_all)
pred_lr = linear_reg.predict(X_all)

price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)
```

```python
plt.yticks(fontname='Arial')
plt.semilogy(data_train.date, data_train.price, label="train data")
plt.semilogy(data_test.date, data_test.price, label="test data")
plt.semilogy(ram_price.date, price_tree, label="tree predict")
plt.semilogy(ram_price.date, price_lr, label="linear predict")
plt.legend()
plt.show()
# 왜 2000년대 이후 tree 성능은 확 떨어지는가?
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
# X[0] = x좌표, X[1] = y좌표, y = classifier(blue or orange)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                   random_state=42)
```

```python
forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("tree {}".format(i))
    mglearn.plots.plot_tree_partition(X, y, tree, ax=ax)

mglearn.plots.plot_2d_separator(forest, X, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("random forest")
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.show()
#각 트리에서 나온 결과를 토대로 다수결로 따져서 최종 결과 도출
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                   random_state=0)
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)
print("train accuracy: {:.3f}".format(forest.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(forest.score(X_test, y_test)))
#결과가 1.0 == overfitting 되었다
```

```python
plt.figure(figsize=(15, 10))
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1, n_features)
    
plot_feature_importances_cancer(forest)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# random: 말 그대로 다양하게 랜덤으로 만들어진다.
# GradientBoost: 이전 트리의 정보를 다음 트리가 반영한다.
# 올라가거나 내려가는 형태로 성능을 향상시켜간다.
```

```python
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)
print("train accuracy: {:.3f}".format(gbrt.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(gbrt.score(X_test, y_test)))
# 각 트리 마다의 depth를 조정하면 overfitting 방지할 수 있다.
```

```python
# hyper_parameter tuning - max_depth
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1) 
gbrt.fit(X_train, y_train)
print("train accuracy: {:.3f}".format(gbrt.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
# hyper_parameter tuning - learning_rate
gbrt = GradientBoostingClassifier(random_state=0, learning_rate=0.01)   
gbrt.fit(X_train, y_train)
print("train accuracy: {:.3f}".format(gbrt.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)

plt.figure(figsize=(15, 10))
plot_feature_importances_cancer(gbrt)
#모델마다 중요시 여기는 특성이 다르기에 모델 선택도 중요하다
```

# Kernel trick
Support Vector Machine(SVM)은 점들이 결정 경계를 support 해준다. 이는 특정한 직선이나 곡선으로 나타내진다.

Kernel trick은 2차원에서 직선 혹은 곡선으로 분류하지 못하는 데이터 셋을 3차원으로 변형시켜 평면으로 분류하는 아이디어다.

이후에 2차원으로 다시 보내면 분류를 해내는 곡선이 하나 생기게 된다.

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state=8)

y = y % 2
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.yticks(fontname='Arial')
plt.xticks(fontname='Arial')
plt.show()
```

```python
from sklearn.svm import LinearSVC

linear_svm = LinearSVC().fit(X, y)
mglearn.plots.plot_2d_separator(linear_svm, X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.show()
```

```python
X_new = np.hstack([X, X[:, 1:]**2])
# 두번째 행을 제곱하여 3번째 열에 추가
from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)

plt.yticks(fontname='Arial')
plt.xticks(fontname='Arial')

mask = y == 0

ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', 
          cmap=mglearn.cm2, s=60, edgecolors='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', 
          cmap=mglearn.cm2, s=60, edgecolors='k')


ax.set_xlabel("feature 0")
ax.set_ylabel("feature 1")
ax.set_zlabel("feature 1 ** 2")

plt.show()
```

```python
np.linspace(1, 10, 20)
```

```python
linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)

plt.yticks(fontname='Arial')
plt.xticks(fontname='Arial')

xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b',
          cmap=mglearn.cm2, s=60, edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', 
          cmap=mglearn.cm2, s=60, edgecolor='k')

ax.set_xlabel("feature0")
ax.set_ylabel("feature1")
ax.set_zlabel("feature1 ** 2")
plt.show()
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contour(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()],
           cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.show()
```

# SVM(Support Vector Machine)

```python
from sklearn.svm import SVC
X, y = mglearn.tools.make_handcrafted_dataset()
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

sv = svm.support_vectors_
sv_labels = svm.dual_coef_.ravel() > 0

mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)

plt.show()
```

```python
# 자료형과 해당 변수가 담당하는 작업, 복수표현을 변수에 표현해주면 가독성이 좋아진다.
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
for ax, C in zip(axes, [-1, 0, 3]):
    for a, gamma in zip(ax, range(-1,2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)
```

## MinMaxScaler
:  0~1 사이의 범위로 축소시킨다.

## StandardScaler
:  평균과 표준편차를 통해 작은 범위로 축소시킨다.

```python
# minmaxscaler:
# (현재 데이터 - 최솟값) / (최대값 - 최솟값)
min_on_training = X_train.min(axis=0)
range_on_trainig = (X_train - min_on_training).max(axis=0)
X_train_scaled = (X_train - min_on_training) / range_on_trainig

print("특성별 최소값 \n", X_train_scaled.min(axis=0))
print("특성별 최대값 \n", X_train_scaled.max(axis=0))
```

```python
X_test_scaled = (X_test - min_on_training) / range_on_trainig
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
print("train acc: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("test acc: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled2 = scaler.transform(X_train)
X_test_scaled2 = scaler.transform(X_test)

svc = SVC()
svc.fit(X_train_scaled2, y_train)
print("train acc: {:.3f}".format(svc.score(X_train_scaled2, y_train)))
print("test acc: {:.3f}".format(svc.score(X_test_scaled2, y_test)))
```
