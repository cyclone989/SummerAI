```python
x = np.random.uniform(-np.pi, np.pi, 500)
```

```python
X = np.array([ [ x[i], y[i] ] for i in range(500)])
```

```python
X.shape
```

```python
sine[:5]
```

```python
model = Sequential()
```

```python
model.add(Dense(10, activation='relu', input_shape=(2,))) # 1층, 10개의 뉴런
model.add(Dense(10, activation='relu')) # 2층
model.add(Dense(10, activation='relu')) # 3층
model.add(Dense(10, activation='relu')) # 4층
model.add(Dense(2, activation='softmax')) # 마지막 층, 다중분류인 softmax
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='sgd')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
names = ['sl', 'sw', 'pl', 'pw', 'class']
```

```python
df = pd.read_csv('datasets/iris.data', names=names)
```

```python
df
```

```python
df.describe()
```

```python
Y_encoded = LabelEncoder().fit_transform(df['class'])
```

```python
Y_encoded
```

```python
Y = np_utils.to_categorical(Y_encoded)
```

```python
X = df.drop('class', axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.1, random_state=0)
```

```python
model = Sequential()
```

```python
model.add(Dense(256, activation='relu', input_shape=(4,)))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train, epochs=50, validation_split=0.1)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.plot(hist.history['acc'], color='green')
plt.plot(hist.history['val_acc'], color='orange')
plt.subplot(122)
plt.plot(hist.history['loss'], color='blue')
plt.plot(hist.history['val_loss'], color='red')
plt.tight_layout()
plt.show()
```

```python
loss_score = model.evaluate(X_test, y_test)
print("loss: {:.2f}".format(loss_score[0]))
print("test acc: {:.2f}".format(loss_score[1]))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
import keras
```

```python
mnist = keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
np.set_printoptions(linewidth=1000)
```

```python
x_train[0]
```

```python
# 0 : black
# 255 : white
```

```python
plt.imshow(x_train[0])
```

```python
plt.imshow(x_train[0], cmap='gray')
```

```python
x_train = x_train / 255.0
```

```python
x_test = x_test / 255.0
```

```python
x_train[0]
```

```python
from keras.layers import Flatten
# 3차원을 2차원으로, 2차원을 1차원으로 차원을 낮춰준다.
```

```python
from keras.layers import Dropout
# Dropout: overfitting 방지
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Flatten(input_shape=(28,28)))    # 28 * 28 ==> 784
model.add(Dense(128, activation='relu'))   
model.add(Dropout(0.2))                    # 20% drop out
model.add(Dense(10, activation='softmax')) # 다중 분류
```

```python
model.compile(loss='sparse_categorical_crossentropy', 
             optimizer='adam', metrics=['accuracy'])
# One Hot 형태로 데이터가 제공되지 않는 경우일 때(like index)
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test, y_test)
# [0.07307151990176644, 0.9777]
# [ loss, accuracy ]
```
