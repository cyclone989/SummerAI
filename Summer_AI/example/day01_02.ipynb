```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.1, random_state=7)
```

```python
import keras
```

```python
from keras.models import Sequential  # ë‡Œ, ê±´ë¬¼
from keras.layers import Dense       # í•œ ì¸µ
```

```python
model = Sequential()
```

```python
t1 = (1,2,3)
print(type(t1))
t2 = (1)
print(type(t2))  # int
t3 = (1,)
print(type(t3))  # tuple
```

```python
model.add(Dense(128, input_shape=(6,), activation='relu')) # 1ì¸µ
model.add(Dense(128, activation='relu'))  # 2ì¸µ
model.add(Dense(1, activation='sigmoid')) # 3ì¸µ
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
%%time
hist = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                epochs=500)
```

```python
plt.figure(figsize=(12,8))
plt.plot(hist.history['loss'])      # í›ˆë ¨ ë¹„ìš©
plt.plot(hist.history['val_loss'])  # ê²€ì¦ ë¹„ìš©
plt.plot(hist.history['acc'])       # í›ˆë ¨ ì •í™•ë„
plt.plot(hist.history['val_acc'])   # ê²€ì¦ ì •í™•ë„
plt.legend(['loss','val_loss','acc','val_acc'])
plt.show()
```
